```
import csv

data =[]

with open ('property_value_small.csv') as csvfile:
    csvdata=csv.reader(csvfile, delimiter=',')
    next(csvdata)
    for row in csvdata:
        data.append({"x": int(row[0]), "y":int(row[1])})
        
print(data)
```

```
import matplotlib.pyplot as plt

for i in range(0, len(data)):
    plt.plot(data[i]["x"], data[i]["y"], 'ro')
```

```
import numpy as np

learningRate=0.000001

def gradientDecent(beta_0, beta_1):
    beta_0_gradient=0
    beta_1_gradient=0
    N=int(len(data))
    for i in range(0,N):
        beta_0_gradient += (((beta_1*data[i]["x"]) + beta_0) - data [i]["y"])
        beta_1_gradient += data[i]["x"]*(((beta_1*data[i]["x"])+beta_0)-data[i]["y"])
    beta_0_new = beta_0 - (learningRate*(2/N)*beta_0_gradient)
    beta_1_new = beta_1 - (learningRate*(2/N)*beta_1_gradient)
    return[beta_0_new, beta_1_new]

def calculateError(beta_0, beta_1):
    errorSum=0
    for i in range (0, len(data)):
        errorSum += np.square(data[i]["y"]-(beta_0 + beta_1*data[i]["x"]))
        
    errorMean=(1/len(data))*errorSum
    return errorMean
```

```
beta_0=0
beta_1=-100

error=calculateError(beta_0, beta_1)
print("Error is "+str(error))

for i in range(0, len(data)):
    plt.plot(data[i]["x"], data[i]["y"], 'ro')
f = lambda x: beta_0 + beta_1*x
x = np.array([0,800])
plt.plot(x,f(x))
plt.show()

for i in range (0,10):
    print("Running gradient descent")
    newBetas = gradientDecent(beta_0, beta_1)
    beta_0 = newBetas[0]
    beta_1 = newBetas[1]
    print("New beta_0 " + str(beta_0)+", new beta_1 = " + str (beta_1))
    error = calculateError(beta_0, beta_1)
    print("Error:" + str(error))
    
    for i in range(0, len(data)):
        plt.plot(data[i]["x"], data[i]["y"], 'ro')
    f = lambda x: beta_0 + beta_1*x
    x = np.array([0,800])
    plt.plot(x,f(x))
    plt.show()
    
```
